```python
import cv2
import matplotlib.pyplot as plt
import os

# Bước 1: Đọc ảnh màu từ file
image = cv2.imread('images/parrot.png')

# Bước 2: Chuyển ảnh màu sang ảnh đen trắng (grayscale)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Bước 3: Lưu ảnh đen trắng vào thư mục chỉ định
output_folder = 'output_images'
output_path = os.path.join(output_folder, 'chuong_1(1).png') 
cv2.imwrite(output_path, gray_image)  

# Bước 4: Hiển thị ảnh gốc và ảnh đen trắng
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Hiển thị ảnh
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Ảnh Gốc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh Đen Trắng')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Bc 1: c nh mu t file
image = cv2.imread('images/parrot.png')

# Bc 2: Chuyn nh mu sang nh en trng (grayscale)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Bc 3: Lu nh en trng vo th mc ch nh
output_folder = 'output_images'
output_path = os.path.join(output_folder, 'chuong_1(1).png') 
cv2.imwrite(output_path, gray_image)  

# Bc 4: Hin th nh gc v nh en trng
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Hin th nh
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('nh Gc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('nh en Trng')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, data, img_as_float
from skimage.restoration import denoise_tv_chambolle
import os

# c nh gc v chuyn sang grayscale
image = color.rgb2gray(io.imread('images/ronaldo.jpg')) 

# To nh nhiu bng cch thm nhiu Gaussian vo nh gc
noisy_img = image + 0.5 * image.std() * np.random.randn(*image.shape)

# Hin th nh gc v nh nhiu
plt.figure(figsize=(12, 9))

# Hin th nh gc
plt.subplot(221)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image', size=20)

# Hin th nh nhiu
plt.subplot(222)
plt.imshow(noisy_img, cmap='gray')
plt.axis('off')
plt.title('Noisy Image', size=20)

# p dng Total Variation Denoising vi trng s = 0.1
denoised_img = denoise_tv_chambolle(noisy_img, weight=0.1)

# Hin th nh  loi b nhiu vi trng s = 0.1
plt.subplot(223)
plt.imshow(denoised_img, cmap='gray')
plt.axis('off')
plt.title('Denoised (weight=0.1)', size=20)

# p dng Total Variation Denoising vi trng s = 1
denoised_img = denoise_tv_chambolle(noisy_img, weight=1)

# Hin th nh  loi b nhiu vi trng s = 1
plt.subplot(224)
plt.imshow(denoised_img, cmap='gray')
plt.axis('off')
plt.title('Denoised (weight=1)', size=20)

# Hin th tt c nh
plt.tight_layout()
plt.show()

# Lu nh kt qu vo th mc ch nh
output_folder = 'output_images'  # Th mc lu nh kt qu

output_path_1 = os.path.join(output_folder, 'chuong_12_denoised_weight_0.1(3).png')  # ng dn lu nh vi weight=0.1
output_path_2 = os.path.join(output_folder, 'chuong 12_denoised_weight_1(3).png')  # ng dn lu nh vi weight=1

# Lu nh kt qu
plt.imsave(output_path_1, denoised_img)  # Lu nh vi trng s 0.1
plt.imsave(output_path_2, denoised_img)  # Lu nh vi trng s 1
```

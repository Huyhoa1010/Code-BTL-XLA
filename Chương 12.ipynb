```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, data, img_as_float
from skimage.restoration import denoise_tv_chambolle
import os

# Đọc ảnh gốc và chuyển sang grayscale
image = color.rgb2gray(io.imread('images/ronaldo.jpg')) 

# Tạo ảnh nhiễu bằng cách thêm nhiễu Gaussian vào ảnh gốc
noisy_img = image + 0.5 * image.std() * np.random.randn(*image.shape)

# Hiển thị ảnh gốc và ảnh nhiễu
plt.figure(figsize=(12, 9))

# Hiển thị ảnh gốc
plt.subplot(221)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image', size=20)

# Hiển thị ảnh nhiễu
plt.subplot(222)
plt.imshow(noisy_img, cmap='gray')
plt.axis('off')
plt.title('Noisy Image', size=20)

# Áp dụng Total Variation Denoising với trọng số = 0.1
denoised_img = denoise_tv_chambolle(noisy_img, weight=0.1)

# Hiển thị ảnh đã loại bỏ nhiễu với trọng số = 0.1
plt.subplot(223)
plt.imshow(denoised_img, cmap='gray')
plt.axis('off')
plt.title('Denoised (weight=0.1)', size=20)

# Áp dụng Total Variation Denoising với trọng số = 1
denoised_img = denoise_tv_chambolle(noisy_img, weight=1)

# Hiển thị ảnh đã loại bỏ nhiễu với trọng số = 1
plt.subplot(224)
plt.imshow(denoised_img, cmap='gray')
plt.axis('off')
plt.title('Denoised (weight=1)', size=20)

# Hiển thị tất cả ảnh
plt.tight_layout()
plt.show()

# Lưu ảnh kết quả vào thư mục chỉ định
output_folder = 'output_images'  # Thư mục lưu ảnh kết quả

output_path_1 = os.path.join(output_folder, 'chuong_12_denoised_weight_0.1(3).png')  # Đường dẫn lưu ảnh với weight=0.1
output_path_2 = os.path.join(output_folder, 'chuong 12_denoised_weight_1(3).png')  # Đường dẫn lưu ảnh với weight=1

# Lưu ảnh kết quả
plt.imsave(output_path_1, denoised_img)  # Lưu ảnh với trọng số 0.1
plt.imsave(output_path_2, denoised_img)  # Lưu ảnh với trọng số 1
```

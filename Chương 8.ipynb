```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, io
from skimage.color import rgb2gray
import os

# Bc 1: c nh v chuyn sang grayscale
img = io.imread('images/zebras.jpg')  # c nh
img_gray = rgb2gray(img)  # Chuyn nh sang grayscale

# Bc 2: p dng Otsu's thresholding  phn on nh
thresh = filters.threshold_otsu(img_gray)  # Tnh ton ngng Otsu
img_binary = img_gray > thresh  # Phn on nh da trn ngng Otsu

# Bc 3: Hin th nh gc v nh  phn on
plt.figure(figsize=(20, 10))

# Hin th nh gc
plt.subplot(131)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.title('Original Image (Grayscale)', size=20)

# Hin th nh sau phn on (binary image)
plt.subplot(132)
plt.imshow(img_binary, cmap='gray')
plt.axis('off')
plt.title('After Otsu Thresholding', size=20)

# Hin th histogram ca nh
plt.subplot(133)
plt.hist(img_gray.ravel(), bins=256, histtype='step', color='black')
plt.axvline(x=thresh, color='red', linestyle='--')
plt.title('Histogram with Otsu Threshold', size=20)

plt.show()

# Bc 4: Lu nh  phn on
output_folder = 'output_images'
output_path_binary = os.path.join(output_folder, 'chuong_8(3).png')
plt.imsave(output_path_binary, img_binary, cmap='gray')  # Lu nh nh phn
```

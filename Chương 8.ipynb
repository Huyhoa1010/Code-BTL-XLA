```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, io
from skimage.color import rgb2gray
import os

# Bước 1: Đọc ảnh và chuyển sang grayscale
img = io.imread('images/zebras.jpg')  # Đọc ảnh
img_gray = rgb2gray(img)  # Chuyển ảnh sang grayscale

# Bước 2: Áp dụng Otsu's thresholding để phân đoạn ảnh
thresh = filters.threshold_otsu(img_gray)  # Tính toán ngưỡng Otsu
img_binary = img_gray > thresh  # Phân đoạn ảnh dựa trên ngưỡng Otsu

# Bước 3: Hiển thị ảnh gốc và ảnh đã phân đoạn
plt.figure(figsize=(20, 10))

# Hiển thị ảnh gốc
plt.subplot(131)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.title('Original Image (Grayscale)', size=20)

# Hiển thị ảnh sau phân đoạn (binary image)
plt.subplot(132)
plt.imshow(img_binary, cmap='gray')
plt.axis('off')
plt.title('After Otsu Thresholding', size=20)

# Hiển thị histogram của ảnh
plt.subplot(133)
plt.hist(img_gray.ravel(), bins=256, histtype='step', color='black')
plt.axvline(x=thresh, color='red', linestyle='--')
plt.title('Histogram with Otsu Threshold', size=20)

plt.show()

# Bước 4: Lưu ảnh đã phân đoạn
output_folder = 'output_images'
output_path_binary = os.path.join(output_folder, 'chuong_8(3).png')
plt.imsave(output_path_binary, img_binary, cmap='gray')  # Lưu ảnh nhị phân
```

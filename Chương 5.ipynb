```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, feature, img_as_float
from skimage.io import imread
from skimage.color import rgb2gray
import pylab
import os
import cv2

# Bước 1: Đọc ảnh và chuyển sang grayscale
img = rgb2gray(imread('images/temple.jpg'))  # Đọc ảnh và chuyển thành grayscale

# Bước 2: Áp dụng bộ lọc Sobel để phát hiện biên
sobel_x = filters.sobel_h(img)  # Đạo hàm theo hướng x (ngang)
sobel_y = filters.sobel_v(img)  # Đạo hàm theo hướng y (dọc)

# Tính tổng bình phương của đạo hàm theo x và y để xác định biên
sobel_edges = np.sqrt(sobel_x**2 + sobel_y**2)

# Bước 3: Lưu ảnh và hiển thị kết quả
output_folder = 'output_images'
output_path_sobel = os.path.join(output_folder, 'chuong_5(3).png')
cv2.imwrite(output_path_sobel, np.uint8(sobel_edges * 255))  # Lưu ảnh biên đã phát hiện

# Bước 4: Hiển thị ảnh gốc và ảnh biên
pylab.figure(figsize=(20, 10))

# Hiển thị ảnh gốc
pylab.subplot(121)
pylab.imshow(img, cmap='gray')
pylab.title('Original Image', size=20)
pylab.axis('off')  # Tắt trục tọa độ

# Hiển thị ảnh biên sau khi áp dụng Sobel Operator
pylab.subplot(122)
pylab.imshow(sobel_edges, cmap='gray')
pylab.title('Edges Detected by Sobel', size=20)
pylab.axis('off')  # Tắt trục tọa độ

pylab.show()
```

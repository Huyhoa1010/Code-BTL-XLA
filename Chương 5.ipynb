```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, feature, img_as_float
from skimage.io import imread
from skimage.color import rgb2gray
import pylab
import os
import cv2

# Bc 1: c nh v chuyn sang grayscale
img = rgb2gray(imread('images/temple.jpg'))  # c nh v chuyn thnh grayscale

# Bc 2: p dng b lc Sobel  pht hin bin
sobel_x = filters.sobel_h(img)  # o hm theo hng x (ngang)
sobel_y = filters.sobel_v(img)  # o hm theo hng y (dc)

# Tnh tng bnh phng ca o hm theo x v y  xc nh bin
sobel_edges = np.sqrt(sobel_x**2 + sobel_y**2)

# Bc 3: Lu nh v hin th kt qu
output_folder = 'output_images'
output_path_sobel = os.path.join(output_folder, 'chuong_5(3).png')
cv2.imwrite(output_path_sobel, np.uint8(sobel_edges * 255))  # Lu nh bin  pht hin

# Bc 4: Hin th nh gc v nh bin
pylab.figure(figsize=(20, 10))

# Hin th nh gc
pylab.subplot(121)
pylab.imshow(img, cmap='gray')
pylab.title('Original Image', size=20)
pylab.axis('off')  # Tt trc ta 

# Hin th nh bin sau khi p dng Sobel Operator
pylab.subplot(122)
pylab.imshow(sobel_edges, cmap='gray')
pylab.title('Edges Detected by Sobel', size=20)
pylab.axis('off')  # Tt trc ta 

pylab.show()
```
